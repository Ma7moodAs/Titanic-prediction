```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print('Setup completed successfully.')
```

```python
array = np.arange(1,11)
print(array**2)

plt.Figure(figsize=(10,6),dpi=100)
plt.plot(array,array**2)
plt.tight_layout()
plt.show()
```

```python
array1 = np.arange(-361,361,10)
plt.figure(figsize=(10,6),dpi=100)
sns.lineplot(x=array1,y=np.sin(np.radians(array1)))
plt.tight_layout()
plt.show()
```
